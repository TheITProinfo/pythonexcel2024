```python
import os
import matplotlib.pyplot as plt
import pandas as pd
path=r"d:\code\pythoncode\pythonexcel\day4"
os.chdir(path)
df=pd.read_excel("salestotal.xlsx")
x=df['Month']
y1=df['Sales']
y2=df['Profit']
plt.bar(x,y1,color='b',width=0.5,label='sales') # width=0.5label='sales'
plt.legend(loc="upper left",fontsize=10) 

plt.grid(True,color='r',axis='y') # 


plt.twinx() # 

plt.plot(x, y2, color='r', marker='o', label='profit') # marker='o'
# plt.ylim(0, 1500000) # y
plt.legend(loc="upper right",fontsize=10) 
plt.title(label="Sales by Month", fontsize=12, fontweight='bold', color='red',loc="left")
plt.xlabel("Month", fontsize=10, fontweight='bold', color='blue', loc="center",labelpad=20)
plt.ylabel("Sales", fontsize=10, fontweight='bold', color='blue', loc="center", labelpad=20)

for a,b in zip(x,y2):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)

plt.show()
```
