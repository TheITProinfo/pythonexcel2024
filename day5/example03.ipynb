```python
import os
import pandas as pd
import xlwings as xw
import matplotlib.pyplot as plt
import numpy as np

path=r"d:\code\pythoncode\pythonexcel\day5"
os.chdir(path)
df=pd.read_excel("salesbyproduct.xlsx")
print(df)
figure=plt.figure()
x=df["product"]
y=df["quantity"]
# print(x)
# print(y)
# plt.pie(y,labels=x,labeldistance=1.1,autopct="%1.2f%%",pctdistance=0.8,startangle=90,radius=1.0,explode=[0,0.5,0,0,0.3,0,0,0,0])
plt.pie(y,labels=x,labeldistance=1.1,autopct="%1.2f%%",pctdistance=0.8,wedgeprops={'width':0.5,'linewidth':2,'edgecolor':'white'})

plt.axis("equal")
plt.title("sales by product",fontdict={'color':'blue','size':'15'},loc='center')
plt.show()
figure.savefig("sales.png")
app=xw.App(visible=False,add_book=False)
workbook=app.books.open("salestotal.xlsx")
worksheet=workbook.sheets[1]
worksheet.pictures.add(figure, name="sales", update=True, left=200, top=0)
workbook.save("saleschart03.xlsx")
workbook.close()
app.quit()
os.remove("sales.png")


```
