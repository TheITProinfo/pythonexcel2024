```python
import os
import pandas as pd
import matplotlib.pyplot as plt # 
import xlwings as xw
path=r"d:\code\pythoncode\pythonexcel\day5"
os.chdir(path)
df=pd.read_excel("salestotal.xlsx")
# print(df)
figure=plt.figure() #创建画布
x=df["Month"]
y=df["Sales"]
plt.plot(x,y,color='red',linewidth=2,linestyle='solid')
plt.title(label='sales by month', loc='center', fontdict={'color': 'blue', 'size': '15'})

max1=df["Sales"].max() # 最大值
min1=df["Sales"].min() # 最小值
# print(max1,min1)
df_max=df[df["Sales"]==max1] # 找到最大值所在的行
df_min=df[df["Sales"]==min1] # 找到最小值所在的行
# print(df_max)
for a,b in zip(df_max["Month"],df_max["Sales"]):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=10) # 给max点添加文字

plt.axis('off') # 关闭坐标轴
plt.grid(True) # 显示网格线
plt.show()
figure.savefig("sales.png") # 保存图片
app=xw.App(visible=False,add_book=False) # load excel apps
workbook=app.books.open("salestotal.xlsx") # 打开excel
worksheet=workbook.sheets[0] # 打开sheet
worksheet.pictures.add(figure, name="sales", update=True, left=200, top=0) # 添加图片 to excel
workbook.save("saleschart02.xlsx")
workbook.close()
app.quit() # 退出excel
os.remove("sales.png") # 删除图片


```
